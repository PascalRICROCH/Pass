```python
wiki_data = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')  # Read the table from Wikipedia
data = wiki_data[0]
```

```python
# prompt: find the stocks values of all companies in S&P 500 from 2022 to today and create a .csv file called 'S&P500_Stock_Values_new.csv'

import pandas as pd
import yfinance as yf

tickers = data['Symbol'].tolist()
start_date = '2022-01-01'
end_date = pd.Timestamp.today().strftime('%Y-%m-%d')

all_stock_data = []
for ticker in tickers:
  try:
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    stock_data['Ticker'] = ticker
    all_stock_data.append(stock_data)
  except:
    print(f"Failed to download data for {ticker}")

combined_data = pd.concat(all_stock_data)
combined_data.to_csv('S&P500_Stock_Values_new.csv')
```

```python
combined_data
```

```python
# prompt: replace frame 'Ticker' by 'Symbol' in combined_data frame

combined_data = combined_data.rename(columns={'Ticker': 'Symbol'})
combined_data
```

```python
# prompt: replace old 'S&P500_Stock_Values_new.csv' with the content of combined_data

combined_data.to_csv('S&P500_Stock_Values_new.csv', mode='w', index=False)
```

```python
# prompt: make sure on 'S&P500_Stock_Values_new.csv' that 'Date' is in the head frame

combined_data = combined_data.reset_index()
combined_data.to_csv('S&P500_Stock_Values_new.csv', mode='w', index=False)
```

```python
# prompt: add a column 'year' in the frame head of the 'S&P500_Stock_Values_new.csv' file that is the 'YEAR' out of the 'Date'

import pandas as pd
combined_data['year'] = pd.to_datetime(combined_data['Date']).dt.year
combined_data.to_csv('S&P500_Stock_Values_new.csv', mode='w', index=False)
```
